```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://cocl.us/datascience_survey_data', index_col = 0)
df.head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
df.sort_values(by=['Very interested'], ascending = False, inplace= True)
df.head(10)
```

```python
df['Very interested'] = (df['Very interested']/2233)*100
df['Somewhat interested'] = (df['Somewhat interested']/2233)*100
df['Not interested'] = (df['Not interested']/2233)*100
```

```python
df.head(10)
```

```python
df['Very interested'] = df['Very interested'].round(decimals=2)
df['Somewhat interested'] = df['Somewhat interested'].round(decimals=2)
df['Not interested'] = df['Not interested'].round(decimals=2)
```

```python
df.head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
ax1 = df.plot(kind='bar', figsize = (20,8), width=0.8, color = ('#5cb85c', '#5bc0de', '#d9534f'), fontsize = 14)
plt.title("Percentage of Respondents' Interest in Data Science Areas", fontsize = 16)
ax1.legend(fontsize = 14)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.spines['left'].set_visible(False)

for i in ax1.patches:
     ax1.text(i.get_x()+0.05, i.get_height()+.5, \
            str(i.get_height()))
```
